```python
import read_deeplift
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
deeplift = read_deeplift.get_deep('deeplift_out/')
```

```python
lengths = []
for task, chroms in deeplift.items():
    for chrom, positions in chroms.items():
        for position, score in positions.items():
            lengths += [len(score)]
print len(lengths)
plt.hist(lengths)
```

```python
one, two = [], []
diff = []
scores = []
for task, chroms in deeplift.items():
    for chrom, positions in chroms.items():
        for position, score in positions.items():
            if len(score) > 20:
                one += [sum(score[0:10])]
                two += [sum(score[10:20])]
                diff += [sum(score[0:10]) / 10.0 - sum(score[10:20]) / 10.0]
            scores += [sum(score) / float(len(score))]
            
plt.scatter(one, two)
plt.show()
plt.hist(diff, bins = 200)
plt.show()
plt.hist(scores, bins = 200)
plt.show()
```
